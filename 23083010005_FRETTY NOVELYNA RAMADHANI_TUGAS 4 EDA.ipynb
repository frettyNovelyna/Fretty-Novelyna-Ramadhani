### Mengatasi Outlier

```python
import pandas as pd

def remove_outliers_iqr(data, column):
    # Menghitung IQR
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

titanic_cleaned = remove_outliers_iqr(titanic, 'fare')
titanic_cleaned = remove_outliers_iqr(titanic_cleaned, 'age')

print(titanic_cleaned.head(10))  
```

Fungsi itu digunakan untuk mengatasi outlier dengan menggunakan metode IQR.

### Kategorisasi Outlier

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


bins_age = [0, 12, 18, 35, 60, 100]
labels_age = ['Anak-anak', 'Remaja', 'Dewasa', 'Dewasa Tua', 'Lansia']
titanic['Age_Category'] = pd.cut(titanic['age'], bins=bins_age, labels=labels_age, right=False)

bins_fare = [0, 10, 30, 100, 600]
labels_fare = ['Rendah', 'Sedang', 'Tinggi', 'Sangat Tinggi']
titanic['Fare_Category'] = pd.cut(titanic['fare'], bins=bins_fare, labels=labels_fare, right=False)

titanic['Pclass_Category'] = titanic['pclass'].map({1: 'Kelas 1', 2: 'Kelas 2', 3: 'Kelas 3'})

print(titanic[['age', 'Age_Category', 'fare', 'Fare_Category', 'pclass', 'Pclass_Category']].head())

plt.figure(figsize=(8, 4))
ax = sns.countplot(data=titanic, x='Age_Category', palette='pastel')
plt.title('Distribusi Kategori Umur')

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom')

plt.show()

plt.figure(figsize=(8, 4))
ax = sns.countplot(data=titanic, x='Fare_Category', palette='pastel')
plt.title('Distribusi Kategori Tarif')

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom')

plt.show()

plt.figure(figsize=(8, 4))
ax = sns.countplot(data=titanic, x='Pclass_Category', palette='pastel')
plt.title('Distribusi Kategori Kelas Penumpang')

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom')

plt.show()
```

Dari kode diatas dikelompokkan berdasarkan tiga kriteria utama yaitu umur, tarif dan kelas penumpang, untuk umur dikelompokkan dalam kategori 'Anak-anak, remaja, dewasa, dewasa tua dan lansia' lalu tarif dikategorikan 'renddah, sedang, tinggi dan sangat tinggi' dan yang terakhir kelas penumpang dikelompokkan 'kelas 1, kelas 2 dan kelas 3'

### Transformasi Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('D:/KULIAH/SEMESTER 3/EDA/data_dummy_Titanic_baru.csv')

numeric_columns = ['Age', 'Fare', 'SibSp', 'Parch']

# Visualisasi Sebelum Transformasi
plt.figure(figsize=(12, 8))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(2, 2, i) 
    sns.histplot(df[column], kde=True, color='skyblue') 
    plt.title(f'{column} - Before Scaling')
plt.tight_layout()
plt.show()

# Transformasi Data dengan StandardScaler
scaler = StandardScaler()
df_scaled = df.copy()  
df_scaled[numeric_columns] = scaler.fit_transform(df[numeric_columns]) 

# Visualisasi Setelah Transformasi
plt.figure(figsize=(12, 8))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df_scaled[column], kde=True, color='salmon')
    plt.title(f'{column} - After Scaling')
plt.tight_layout()
plt.show()
```

Sebelum transformasi data ditampilkan menggunakan histogram dengan garis KDE(Kernel Density Estimation) untuk melihat bagaimana distribusi kolom Age,fare, SIBSP dan parch. Lalu setelah transformasi histogram baru dengan KDE ditampilkan untuk dapat melihat peruubahan distribusi data, namun meskipun bentuk distribusi tidak berubah secara signifikan.
